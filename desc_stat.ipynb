```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

desc_stat = pd.read_csv('ED_study_stat.csv')

def meanSort(title):
    title = desc_stat[title]
    a_rray = np.array(title)

    LDlist = list()
    for i in range(0,len(a_rray)):
        if i%3 == 0:
            LDlist.append(a_rray[i])
    return(LDlist)

array1 = meanSort('Loneliness Difference\n(T3 - T1)')
array2 = meanSort('Perceived Social Support Difference (T3 - T1)')
array3 = meanSort('Savoring Habits Difference (T3 - T1)')

# Names for the columns
categories = ['Parent', 'No Parent','Sibling','No Sibling','Family','No Family','Friend','No Friend','Romantic','No Romantic']
x = np.arange(len(categories))
bar_width = 0.25
# Show plot

fig = plt.figure(figsize=(10,6))   #creates a blank canvas 
ax = fig.add_axes([1,1,1,1]) #creates an axes - (starting pt x, starting pt y, fractional size x, fractional size y)
ax.bar(x - bar_width, array1, bar_width, label = 'Loneliness Difference (T3 - T1)',color = 'green')
ax.bar(x,array2,bar_width,label='Perceived Social Support Difference (T3 - T1)',color='red')
ax.bar(x + bar_width,array3,bar_width,label='Savoring Habits Difference (T3 - T1)',color='blue')
ax.set_xlabel('Type of Relationship')
ax.set_ylabel('Difference in Mean')
ax.grid(True)
ax.legend()
ax.set_title('Difference in Means Between Measures')
plt.xticks(x, categories, rotation=45)
plt.show()

```
